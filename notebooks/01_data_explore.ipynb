```python

def sort_columns_excluding_first_n(df, n):
    """
    Sort all but the first n columns of a DataFrame by their column names.

    Parameters:
    df (pd.DataFrame): The original DataFrame.
    n (int): The number of first columns to exclude from sorting.

    Returns:
    pd.DataFrame: A new DataFrame with all but the first n columns sorted by column name.
    """

    # Step 1: Split the DataFrame
    excluded_part = df.iloc[:, :n]  # First n columns to exclude from sorting
    sortable_part = df.iloc[:, n:]  # The rest of the columns

    # Step 2: Sort the rest of the columns by their names
    sorted_rest = sortable_part[sorted(sortable_part.columns, key=lambda x: int(x))]

    # Step 3: Reassemble the DataFrame
    df_sorted = pd.concat([excluded_part, sorted_rest], axis=1)

    return df_sorted



def move_column_to_first(df, column_name):
    """
    Move a specified column by name to the first position in the DataFrame.

    Parameters:
    df (pd.DataFrame): The original DataFrame.
    column_name (str): The name of the column to move.

    Returns:
    pd.DataFrame: A new DataFrame with the specified column moved to the first position.
    """

    # Ensure the column exists in the DataFrame
    if column_name in df.columns:
        # Create a list of columns with the specified column at the first position
        cols = [column_name] + [col for col in df.columns if col != column_name]

        # Reindex the DataFrame with the new column order
        return df[cols]
    else:
        # Return the original DataFrame if the column name doesn't exist
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return df
```

### Compare 2 Voest Datasets

```python
# Performing an inner join
voest_ds_1["source_df1"] = True
voest_ds_2["source_df2"] = True

inner_join_df = pd.merge(voest_ds_1[["BALANCING_TIME_UNIT_UTC"]+col_names2], voest_ds_2[["BALANCING_TIME_UNIT_UTC"]+col_names2], how='inner', on='BALANCING_TIME_UNIT_UTC', suffixes=('_df1', '_df2'))

# Performing an outer join
outer_join_df = pd.merge(voest_ds_1[["BALANCING_TIME_UNIT_UTC", "source"]+col_names2], voest_ds_2[["BALANCING_TIME_UNIT_UTC", "source"]+col_names2], how='outer', on='BALANCING_TIME_UNIT_UTC', suffixes=('_df1', '_df2'))

# Calculating the difference in rows
rows_left_out = len(outer_join_df) - len(inner_join_df)

missing_from_df1 = outer_join_df[outer_join_df['source_df1'].isna()]
missing_from_df2 = outer_join_df[outer_join_df['source_df2'].isna()]

missing_count_df1 = len(missing_from_df1)
missing_count_df2 = len(missing_from_df2)

print(f"Rows exclusive to df2 (missing from df1): {missing_count_df1}")
print(f"Rows exclusive to df1 (missing from df2): {missing_count_df2}")

print(f"Number of rows left out when switching from an outer join to an inner join: {rows_left_out}")
```

```python
matches = {}

def compare_with_nan(col1, col2):
    # If both are NaN, consider it a match
    # If one is NaN, also consider it a match
    return (col1 == col2) | col1.isna() | col2.isna()


for column in col_names2:
    col_df1 = f'{column}_df1'
    col_df2 = f'{column}_df2'

    # Checking if the columns match
    #comparison_column = f'{column}_match'
    #comparison_column_nan = f'{column}_matchnan'
    comparison_column = inner_join_df[col_df1] == inner_join_df[col_df2]
    comparison_column_nan = compare_with_nan(inner_join_df[col_df1], inner_join_df[col_df2])

    
    # Counting matches
    matches[column] = (comparison_column.sum(), comparison_column_nan.sum())

for column, match_count in matches.items():
    print(f"Number of matching entries for {column}: {match_count}/{len(inner_join_df)}")
```

```python
threshold = 1e-6
nan_threshhold = 0.1

columns_to_drop = [col for col in voest_ds_2.columns[5:] if voest_ds_2[col].std() < threshold or voest_ds_2[col].isna().mean() > nan_threshhold]

voest_ds_2_dropped = voest_ds_2.drop(columns=columns_to_drop)
# Forward fill the remaining NaNs
voest_ds_2_dropped.ffill(inplace=True)
print(f"Dropped {len(columns_to_drop)}")
for col in columns_to_drop:
    print(f"Column '{col}' has a standard deviation of {voest_ds_2[col].std()} and a NaN ratio of {voest_ds_2[col].isna().mean()}")
print(f"Remaining columns: ")

print(voest_ds_2_dropped[5:].nunique()<10)
voest_ds_2_dropped
```

```python
voest_ds_2_dropped.iloc[:, 5:] = (voest_ds_2_dropped.iloc[:, 5:] - voest_ds_2_dropped.iloc[:, 5:].mean()) / voest_ds_2_dropped.iloc[:, 5:].std()
voest_ds_2_dropped["PROGNOSE-EXT_Preise_EURspez_AE00-ENTSOE-Indikative"] = (voest_ds_2_dropped["PROGNOSE-EXT_Preise_EURspez_AE00-ENTSOE-Indikative"] - voest_ds_2_dropped["PROGNOSE-EXT_Preise_EURspez_AE00-ENTSOE-Indikative"].mean()) / voest_ds_2_dropped["PROGNOSE-EXT_Preise_EURspez_AE00-ENTSOE-Indikative"].std()
```
